```python
# An in-memory TfidfRetriever based on Pandas dataframes

# from haystack.nodes import TfidfRetriever
from haystack.nodes import ElasticsearchRetriever,TfidfRetriever

# retriever = ElasticsearchRetriever(document_store=document_store)
retriever = TfidfRetriever(document_store=document_store)
```

```python
reader = FARMReader(model_name_or_path="deepset/bert-base-cased-squad2", use_gpu=True,)
```

```python
from haystack.pipelines import ExtractiveQAPipeline,MostSimilarDocumentsPipeline,

pipe = ExtractiveQAPipeline(reader, retriever)

# pipe_2 = MostSimilarDocumentsPipeline(reader, retriever)
```

```python
# You can configure how many candidates the reader and retriever shall return
# The higher top_k for retriever, the better (but also the slower) your answers.
prediction = pipe.run(
    query="Activation Function", params={"Retriever": {"top_k": 10}, "Reader": {"top_k": 10}}
)
```
